### clustering on inferCNV matices

```r
all_mtx<- readRDS("./data/infercnv/mtx_all.rds")
obj<- CreateSeuratObject(counts = all_mtx, project = "infercnv", min.cells = 0, min.features = 0)
obj@meta.data$genotype<- B_cell@meta.data[rownames(obj@meta.data), "genotype"]
obj@meta.data$age<- B_cell@meta.data[rownames(obj@meta.data), "age"]
obj@meta.data$pool_id<- B_cell@meta.data[rownames(obj@meta.data), "pool_id"]
```

```r
obj<- FindVariableFeatures(obj, selection.method = "vst", nfeatures = 2000)
obj<- ScaleData(obj)
obj <- RunPCA(obj)
ElbowPlot(obj,ndims = 50)
obj <- FindNeighbors(obj, dims = 1:15)
obj <- FindClusters(obj, resolution = 0.6)
obj <- RunUMAP(obj, dims = 1:15)
saveRDS(obj, "./data/infercnv/inferCNV_mtx_all_obj_res0.6_pc15.rds")
```

### genotype enrichment in inferCNV define clusters

```r
obj <- readRDS("./data/infercnv/inferCNV_mtx_all_obj_res0.6_pc15.rds")
```

```r
DimPlot(obj, reduction = "umap", group.by = "genotype")
ggsave("./results/figures/10_inferCNV_umap_by_genotype.pdf", width = 5, height = 4)
DimPlot(obj, reduction = "umap", label = TRUE)
ggsave("./results/figures/10_inferCNV_umap_by_clusterid.pdf", width = 5, height = 4)
```

```r
library(Polychrome)
col_4 <- kelly.colors(6)[-1]
names(col_4) = NULL
df<- obj@meta.data  %>% 
    mutate(age = factor(age, levels = c("6mos", "14mos", "18mos", "sick")))  %>% 
    mutate(genotype = factor(genotype, levels = c("WT", "CD70-/-", "Bcl6tg/+", "CD70-/-;Bcl6tg/+")))  %>% 
    count(seurat_clusters, genotype, name = "n_cells")  %>% 
    left_join(obj@meta.data  %>% count(seurat_clusters, name = "total"), by = "seurat_clusters")  %>% 
    mutate(percentage = n_cells/total)

print(df  %>% filter(genotype == "WT")  %>% arrange(percentage))
clusterid_fact<- df  %>% filter(genotype == "WT")  %>% arrange(percentage)  %>% pull(seurat_clusters)

df  %>% 
    mutate(seurat_clusters = factor(seurat_clusters, levels = clusterid_fact))  %>% 
    ggplot(aes(x = seurat_clusters, y = percentage, fill = genotype)) +
    geom_bar(stat = "identity", position = 'stack') +
    scale_y_continuous(labels = scales::percent, limits = c(0, 1)) +
    scale_fill_manual(values = col_4) +
    theme_bw() +
    theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) +
    labs(x = "Cluster", y = "Percentage of cells per cluster", fill = "genotype")

ggsave("./results/figures/10_inferCNV_barplot_percentage_of_genotypes_per_cluster.pdf", width = 8, height = 6)
```

#### cutoff of 0.1 is chosed based on enrichment of WT cells in each cluter

```r
malignant_clusters<- df  %>% filter(genotype== "WT")  %>% filter(percentage < 0.1)  %>% pull(seurat_clusters)
malignant_clusters
malignant_cells<- rownames(obj@meta.data  %>% filter(seurat_clusters %in% malignant_clusters) )
```

### store inferCNV identified malignant cells in B_cell object

```r
B_cell@meta.data[, "inferCNV.results"] = "non-malignant"
B_cell@meta.data[malignant_cells, "inferCNV.results"] = "malignant"
B_cell@meta.data[rownames(B_cell@reductions$umap@cell.embeddings), "UMAP_1"] = B_cell@reductions$umap@cell.embeddings[,'UMAP_1']
B_cell@meta.data[rownames(B_cell@reductions$umap@cell.embeddings), "UMAP_2"] = B_cell@reductions$umap@cell.embeddings[,'UMAP_2']
```

```r
B_cell@meta.data  %>% 
    mutate(age = factor(age, levels = c("6mos", "14mos", "18mos", "sick")))  %>% 
    mutate(genotype = factor(genotype, levels = c("WT", "CD70-/-", "Bcl6tg/+", "CD70-/-;Bcl6tg/+")))  %>%
    ggplot(aes(x = UMAP_1, y = UMAP_2)) +
    geom_point(aes(col = inferCNV.results), size = 0.2) +
    labs(x = "UMAP1", y = "UMAP2", title = "Malignant cells in B-cell Space", col = "InferCNV Predictions") +
    scale_color_manual(values = c("deepskyblue3", "gray")) +
    theme_bw(base_size = 14)  +
    guides(color = guide_legend(override.aes = list(size = 5)))

ggsave("./results/figures/10_inferCNV_denfined_malignant_cells_B_cell_umap.pdf", width = 5, height = 4)
```

```r
B_cell@meta.data  %>% 
    mutate(age = factor(age, levels = c("6mos", "14mos", "18mos", "sick")))  %>% 
    mutate(genotype = factor(genotype, levels = c("WT", "CD70-/-", "Bcl6tg/+", "CD70-/-;Bcl6tg/+")))  %>%
    ggplot(aes(x = UMAP_1, y = UMAP_2)) +
    geom_point(aes(col = inferCNV.results), size = 0.05) +
    labs(x = "UMAP1", y = "UMAP2", title = "Malignant cells in B-cell Space", col = "InferCNV Predictions") +
    scale_color_manual(values = c("deepskyblue3", "gray")) +
    theme_bw(base_size = 14)  +
    facet_grid(age ~ genotype) +
    guides(color = guide_legend(override.aes = list(size = 5)))

ggsave("./results/figures/10_inferCNV_denfined_malignant_cells_B_cell_umap_by_genotype_by_age.pdf", width = 10, height = 8)
```
