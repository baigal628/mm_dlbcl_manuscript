# percentage of infercnv_pos cells per mouse

```r
total_cells_per_mouse =  B_cell@meta.data %>% count(sample_id, genotype, age, name='total_cells_per_mouse')

total_infercnv_pos_cells_per_mouse<- B_cell@meta.data  %>% 
    count(sample_id, infercnv, name='total_infercnv_pos_cells_per_mouse')  %>% 
    complete(sample_id, infercnv, fill = list(total_infercnv_pos_cells_per_mouse = 0))  %>% filter(infercnv == 'infercnv_pos')

total_cells_per_mouse %>%
    left_join(total_infercnv_pos_cells_per_mouse, by = c("sample_id"))  %>% 
    mutate(percent_infercnv_pos_cell_per_mouse = total_infercnv_pos_cells_per_mouse/total_cells_per_mouse)  %>% 
    mutate(percent_infercnv_pos_cell_per_mouse = replace_na(percent_infercnv_pos_cell_per_mouse, 0))  %>% 
    mutate(age = factor(age, levels = c("6mos", "14mos", "18mos", "sick")))  %>% 
    mutate(genotype = factor(genotype, levels = c("WT", "CD70-/-", "Bcl6tg/+", "CD70-/-;Bcl6tg/+")))  %>% 
    ggplot(aes(x=genotype, y = percent_infercnv_pos_cell_per_mouse)) +
    geom_boxplot(aes(fill = genotype), outlier.colour = "NA") +
    geom_point() +
    scale_y_continuous(labels = scales::percent, limits = c(0,1)) +
    stat_compare_means(aes(label = paste0("p = ", ..p.format..)), vjust=0.5) +
    theme_bw(base_size=14)+
    facet_grid(~age, scales = "free_y") +
    theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1), legend.position = "none") +
    labs(x = "Genotype", y = "percentage of infercnv positive\ncells per mouse")

ggsave('./results/figures/12_inferCNV_percent_infercnv_pos_cell_per_mouse.pdf', width = 8, height = 4)
```

```r
df<- total_cells_per_mouse %>%
    left_join(total_infercnv_pos_cells_per_mouse, by = c("sample_id"))  %>% 
    mutate(percent_infercnv_pos_cell_per_mouse = total_infercnv_pos_cells_per_mouse/total_cells_per_mouse)  %>% 
    mutate(percent_infercnv_pos_cell_per_mouse = replace_na(percent_infercnv_pos_cell_per_mouse, 0))  %>% 
    mutate(age = factor(age, levels = c("6mos", "14mos", "18mos", "sick")))  %>% 
    mutate(genotype = factor(genotype, levels = c("WT", "CD70-/-", "Bcl6tg/+", "CD70-/-;Bcl6tg/+")))


comparisons <- list(
  c("WT", "CD70-/-;Bcl6tg/+"),
  c("Bcl6tg/+", "CD70-/-;Bcl6tg/+"),
  c("WT", "Bcl6tg/+"),
  c("CD70-/-", "CD70-/-;Bcl6tg/+")
)

ages <- c("6mos", "14mos", "18mos", "sick")

run_wilcox <- function(thisage, pair) {
    
    group1 <- pair[1]
    group2 <- pair[2]
    test_df <- df %>% 
        filter(age == thisage) %>% 
        select(sample_id, genotype, age, percent_infercnv_pos_cell_per_mouse)
  
    
    data1 <- test_df$percent_infercnv_pos_cell_per_mouse[test_df$genotype == group1]
    data2 <- test_df$percent_infercnv_pos_cell_per_mouse[test_df$genotype == group2]
    
    if (length(data1) > 1 & length(data2) > 1) {
        test_result <- wilcox.test(data1, data2, exact = FALSE)
        tibble(
          age = thisage,
          group1 = group1,
          group2 = group2,
          p_value = test_result$p.value,
          W_stat = test_result$statistic)
    } else {
    tibble(age = thisage, group1 = group1, group2 = group2, p_value = NA, W_stat = NA)
    }
}


stat<- c()
for (thisage in ages){
    for (comp in comparisons){
        this_df<- run_wilcox(thisage, comp)
        stat<- rbind(stat, this_df)
    }
}

stat_results <- stat %>% 
    mutate(age = factor(age, levels = c("6mos", "14mos", "18mos", "sick")))  %>% 
    mutate(x_pos = (match(group1, levels(df$genotype)) + match(group2, levels(df$genotype))) / 2) %>%
    group_by(age) %>% 
    mutate(y_pos = 0.8 + (0.05 * (row_number() - 1))) %>% 
    ungroup()  %>% mutate(p_label = ifelse(p_value < 0.001, "p < 0.001", sprintf("p = %.3f", p_value)))
```

```r
df %>% 
    ggplot(aes(x = genotype, y = percent_infercnv_pos_cell_per_mouse)) +
    geom_boxplot(aes(fill = genotype), outlier.colour = "NA") +
    geom_point() +
    scale_y_continuous(labels = scales::percent, limits = c(0, 1)) +
    theme_bw(base_size = 14) +
    facet_grid(~age, scales = "free_y") +
    theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1), 
          legend.position = "none") +
    labs(x = "Genotype", y = "Percentage of infercnv positive\ncells per mouse") +
    geom_segment(
        data = stat_results  %>% filter(!is.na(p_label)),
        aes(x = match(group1, levels(df$genotype)),
            xend = match(group2, levels(df$genotype)),
            y = y_pos, yend = y_pos),
        inherit.aes = FALSE
      ) +
    geom_text(
        data = stat_results %>% filter(!is.na(p_label)),
        aes(x = x_pos, y = y_pos, label = p_label),
        inherit.aes = FALSE,
        vjust = -0.5,
        size = 3.5
      )


ggsave('./results/figures/12_inferCNV_percent_infercnv_positive_cell_per_mouse_paired.pdf', width = 8, height = 6)
```

##  for each mouse, which cluster the infercnv_pos cells are enriched

```r
library(ggpubr)
library(rstatix)
```

```r
n_infercnv_pos_cells_per_mouse_per_cluster<- 
    B_cell@meta.data  %>% 
    count(sample_id, infercnv, seurat_clusters, name ='n_infercnv_pos_cells_per_mouse_per_cluster')  %>% 
    complete(sample_id, infercnv, seurat_clusters, fill = list(n_infercnv_pos_cells_per_mouse_per_cluster = 0))  %>% 
    filter(infercnv == 'infercnv_pos')  %>% 
    left_join(B_cell@meta.data  %>% distinct(genotype, age, sample_id), by = "sample_id")

df_infercnv_pos<- n_infercnv_pos_cells_per_mouse_per_cluster   %>% 
    left_join(total_infercnv_pos_cells_per_mouse, by = c("sample_id"))  %>% 
    mutate(celltype_enrich = n_infercnv_pos_cells_per_mouse_per_cluster/total_infercnv_pos_cells_per_mouse)  %>% 
    mutate(celltype_enrich = replace_na(celltype_enrich, 0))  %>% 
    mutate(age = factor(age, levels = c("6mos", "14mos", "18mos", "sick")))  %>% 
    mutate(genotype = factor(genotype, levels = c("WT", "CD70-/-", "Bcl6tg/+", "CD70-/-;Bcl6tg/+")))
```

```r
color_p = B_cell@meta.data  %>%
    distinct(seurat_clusters, HEX_code)  %>% 
    arrange(seurat_clusters)  %>% 
    pull(HEX_code)


stat_results <- compare_means(celltype_enrich ~ genotype, 
                              data = df_infercnv_pos %>%
                                          group_by(seurat_clusters, age) %>%
                                          filter(n_distinct(genotype) > 1), 
                              group.by = c("seurat_clusters", "age"), 
                              method = "kruskal.test")

df<- df_infercnv_pos  %>% left_join(stat_results)


df %>%
  arrange(seurat_clusters) %>% 
  ggplot(aes(x = genotype, y = celltype_enrich, color = seurat_clusters)) +
  geom_boxplot() +
  geom_point() +
  theme_bw(base_size = 14) +
  scale_fill_manual(values = color_p) +
  facet_grid(seurat_clusters ~ age, scales = "free_y") +
  scale_y_continuous(labels = scales::percent, limits = c(0, 1)) +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1), legend.position = "none") +
  labs(x= 'age x genotype', y = '# infercnv_pos cells per cluster/total # infercnv_pos cells') +
  geom_text(aes(x = 1, y = 0.95, label = paste0("p = ", round(p, 3))), 
            inherit.aes = FALSE,  # Don't inherit aesthetics
            data = df %>% 
              group_by(seurat_clusters, age) %>% filter(!is.na(p)),
            color = "black", 
            size = 3, 
            hjust = 0,
            vjust = 1) 

ggsave('./results/figures/12_inferCNV_cell_type_enrichment_in_infercnv_pos.pdf', width = 8, height = 20)
```

##  for each mouse, which cluster the infercnv_neg cells are enriched

```r
total_infercnv_neg_per_mouse = B_cell@meta.data  %>%  filter(infercnv=='infercnv_neg')  %>% 
    count(sample_id, genotype, age, name='total_infercnv_neg_cells_per_mouse')

total_cells_per_mouse %>%
    left_join(total_infercnv_neg_per_mouse, by = c("genotype", "age", "sample_id"))  %>% 
    mutate(percent_infercnv_neg_cell_per_mouse = total_infercnv_neg_cells_per_mouse/total_cells_per_mouse)  %>% 
    mutate(percent_infercnv_neg_cell_per_mouse = replace_na(percent_infercnv_neg_cell_per_mouse, 0))  %>% 
    mutate(age = factor(age, levels = c("6mos", "14mos", "18mos", "sick")))  %>% 
    mutate(genotype = factor(genotype, levels = c("WT", "CD70-/-", "Bcl6tg/+", "CD70-/-;Bcl6tg/+")))  %>% 
    ggplot(aes(x=genotype, y = percent_infercnv_neg_cell_per_mouse)) +
    geom_boxplot(aes(fill = genotype), outlier.colour = "NA") +
    geom_point() +
    theme_bw(base_size=14)+
    facet_grid(~age, scales = "free_y") +
    theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1), legend.position = "none") +
    stat_compare_means(aes(label = paste0("p = ", ..p.format..)), vjust=40)

ggsave('./results/figures/12_inferCNV_percent_infercnv_neg_cell_per_mouse.pdf', width = 8, height = 4)
```

```r
df<- B_cell@meta.data
# Fraction of infercnv_pos Cells Per Mouse
mal_fracs = df %>% filter(genotype %in% c("Bcl6tg/+")) %>% 
    count(sample_id, infercnv) %>% group_by(sample_id) %>% 
    mutate(Frac = n/sum(n)) %>% filter(infercnv=="infercnv_pos")

#  Fraction of infercnv_pos Cells in Each Cluster Per Mouse 43
ct_fracs = df %>% filter(genotype %in% c("Bcl6tg/+")) %>% 
    filter(infercnv=="infercnv_pos") %>% count(sample_id, seurat_clusters) %>% 
    group_by(sample_id) %>% mutate(Frac = n/sum(n)) %>% 
    pivot_wider(id_cols = sample_id, names_from = seurat_clusters, values_from = Frac) %>% as.data.frame

rownames(ct_fracs) = ct_fracs[, 1]
ct_fracs = ct_fracs[, -1]
ct_fracs[is.na(ct_fracs)] = 0
# Correlation between clusters enrichment and
ct_mal_cors = cor(ct_fracs, mal_fracs$Frac[match(rownames(ct_fracs), mal_fracs$sample_id)])

#  Fraction of infercnv_neg Cells in Each Cluster Per Mouse 
ct_fracs = df %>% filter(genotype %in% c("Bcl6tg/+")) %>% 
    filter(infercnv =="infercnv_neg") %>% count(sample_id, seurat_clusters) %>% 
    group_by(sample_id) %>% mutate(Frac = n/sum(n)) %>% 
    pivot_wider(id_cols = sample_id, names_from = seurat_clusters, values_from = Frac) %>% as.data.frame

rownames(ct_fracs) = ct_fracs[, 1]
ct_fracs = ct_fracs[, -1]
ct_fracs[is.na(ct_fracs)] = 0

ct_nonmal_cors = cor(ct_fracs, mal_fracs$Frac[match(rownames(ct_fracs), mal_fracs$sample_id)], use = "pairwise.complete.obs")

gg4plot = data.frame(Mal = ct_mal_cors, NonMal = ct_nonmal_cors[rownames(ct_mal_cors), 1], Cluster = rownames(ct_mal_cors))

ggplot(gg4plot, aes(NonMal, Mal)) + 
    geom_point() + geom_abline(linetype="dashed") +
    geom_text_repel(aes(label = Cluster)) + 
    theme_bw(base_size = 14) +
    xlim(-0.5, 0.5) +
    ylim(-0.5, 0.5) +
    labs(x = "Correlation of non infercnv_pos cell enrichment", y = "Correlation of infercnv_pos cell enrichment")

ggsave('./results/figures/12_correlation_of_infercnv_pos_infercnv_neg_cell_enrichment_with_total_n_cells.pdf', 
       width = 6, height = 5)
```
